### **(A) DISCRETE DISTRIBUTIONS**


 ### **(I) Binomial Distribution B(n,p):** <br>
 N independent trials (bernoulli trials) <br> p is the fixed probability of success <br>
 It is the model when we have a sequence of independent binary events (sucess/failure) with same probability of sucess <br>
 image.png<br>
 **Expected Value/Mean** = np<br>
 **Variance** = np(1-p)


```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
n = 30
p = 0.5
u,var,sk,kt = binom.stats(n,p,moments='mvsk')
print("Mean: {0}".format(u))
print("Variance: {0}".format(var))
print("Skewness: {0}".format(sk))
print("Kurtosis: {0}".format(kt))
x = list(range(0,50))
plt.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
plt.show()
```

### **(II) Hypergeometric Distribution H(N,M,n):** <br>

```python
# from scipy docs
from scipy.stats import hypergeom
import numpy as np
import matplotlib.pyplot as plt
M,n,N = 20,7,12
rv = hypergeom(M,n,N)
x = np.arange(0,n+1)
pmf = rv.pmf(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,pmf,'bo')
plt.show()

# to gen random numbers: R = hypergeom.rvs(M, n, N, size=10)
```

**<ins>2.Large Sample Approximations</ins>**
---

**The central limit theorem says:<br>**
The distribution of sample means approximates a normal distribution as the sample size gets larger, no matter what the population distribution. Usually n>30 is sufficient condition.

**The law of large numbers says:<br>**
The result of performing the same experiment a large number of times will result the average of the results obtained from large number of trials should be close to the expected value. It gets closer and closer as number of experiments increase.<br>
